```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('title')
plt.plot(x,y,'r')
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(3,2,2)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range 0 to 1)
axes.plot(x,y,'r--')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])



axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('title')
axes1.plot(x,y,'r--')



axes2.set_xlabel('x axis')
axes2.set_ylabel('y axis')
axes2.set_title('title')
axes2.plot(y,x,'g*-')
```

```python
#tuple unpacking
fig,axes = plt.subplots()

axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('title')
axes.plot(x,y,'r--')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
for ax in axes:
    ax.set_xlabel('x axis')
    ax.set_ylabel('y axis')
    ax.set_title('title')
    ax.plot(x,y,'r--')
    
fig    
```

```python
#fig = plt.figure(figsize=(8,2))
fig,ax = plt.subplots(figsize=(3,2),dpi=150)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('title')
ax.plot(x,y,'r--')
```

```python
fig.savefig("filename.png", dpi=200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x**2,label='x-axis')
axes.plot(x,x**3,label='y-axis')
axes.legend()
```

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'b.-') 
ax.plot(x, x**2, 'g.-',alpha=0.5) 
ax.plot(x, x**3, color="red", linewidth=2.00)
ax.plot(x, x**2.5, color="blue", lw=3.00,linestyle='--')
ax.plot(x, x**2.8, color="black", lw=3.00,ls='--',marker='+',markersize=10.00,markerfacecolor='white')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_ylim([0, 60])
axes[1].set_xlim([2, 5])
axes[1].set_title("tight axes")
```
